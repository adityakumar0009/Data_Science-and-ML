```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,kind='scatter')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='time')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,style='time')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,style='time',hue='smoker')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='time',col='smoker')
plt.show()
```
