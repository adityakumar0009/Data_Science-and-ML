```python
x = df[["Average Cost for two","Has Table booking","Has Online delivery","Price range"]]
```

```python
# Above 2 below 2.5 poor
# Above 2.5 below 3.5 average
# Above 3.5 below 4.0 good
# Above 4.0 below 4.5 very good
# Above 4.5 excellent
```

```python
y = df[["Aggregate rating"]]
```

```python
x["Price range"].value_counts()
```

```python
from sklearn import preprocessing
```

```python
labelencoder =  preprocessing.LabelEncoder()
```

```python
x['Has Table booking'] = labelencoder.fit_transform(x['Has Table booking'])
```

```python
x
```

```python
# has table booking 0 is no and 1 is yes
```

```python
labelencoder.classes_
```

```python
x['Has Online delivery'] = labelencoder.fit_transform(x['Has Online delivery'])
```

```python
x
```

```python
# has online delivery 0 is no and 1 is yes
```

```python
from sklearn.preprocessing import StandardScaler
Scaler = StandardScaler()
```

```python
x = Scaler.fit_transform(x)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
def modelresult(predictions):
    print("mean absolute error on model {}".format(mean_absolute_error(y_test,predictions)))
    print("root mean squared error on model {}".format(np.sqrt(mean_squared_error(y_test,predictions))))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
predictionsfromlr = lr.predict(x_test)
```

```python
modelresult(predictionsfromlr)
```

```python
predictionsfromlr
```

```python
from sklearn.svm import SVR
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
svrmodel = SVR()
```

```python
param_gridsvr = {"C":[0.1,0.5],"kernel":["linear","poly"],"degree":[2]}
```

```python
gridsvr = GridSearchCV(svrmodel,param_gridsvr)
```

```python
gridsvr.fit(x_train,y_train)
```

```python
predsgridsvr = gridsvr.predict(x_test)
```

```python
modelresult(predsgridsvr)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
param_grid = {
    "max_depth" : [10,30],
    "min_samples_leaf" : [1,2],
    "min_samples_split" : [2,5]
}
```

```python
treemodel =  DecisionTreeRegressor()
```

```python
grid_tree = GridSearchCV(estimator=treemodel,param_grid=param_grid)
```

```python
grid_tree.fit(x_train,y_train)
```

```python
treeprediction = grid_tree.predict(x_test)
```

```python
modelresult(treeprediction )
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfmodel = RandomForestRegressor()
```

```python
parm_gridrfr = {"max_depth" : [5,15],"n_estimators":[2,5,10]}
```

```python
gridrfr =  GridSearchCV(rfmodel,parm_gridrfr )
```

```python
gridrfr.fit(x_train,y_train)
```

```python
randomforestpredictions = gridrfr.predict(x_test)
```

```python
modelresult(randomforestpredictions)
```

```python
gridrfr.best_params_
```

```python
grid_tree.best_params_
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
KNN = KNeighborsRegressor()
```

```python
knn_param_grid = {"n_neighbors" : [9,11,13,15,17]}
```

```python
knn_grid_search = GridSearchCV(KNN,knn_param_grid)
```

```python
knn_grid_search.fit(x_train,y_train)
```

```python
knnpreds = knn_grid_search.predict(x_test)
```

```python
modelresult(knnpreds)
```

```python
knn_grid_search.best_params_
```

```python
from sklearn.ensemble import AdaBoostRegressor
```

```python
ada = AdaBoostRegressor()
```
