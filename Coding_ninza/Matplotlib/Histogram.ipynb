```python
b = [1,2,3,4,1,2,3,4,5,6,7,10,11,12,15,13,14,12,5,5,7,16,17,16,18,19]
```

```python
import numpy as np
```

```python
xt = np.arange(23)
xt
```

```python
plt.hist(b,edgecolor="black")
plt.xticks(xt)
plt.show()
```

```python
plt.hist(b,bins=21,edgecolor="black")
plt.xticks(xt)
plt.show()
```

```python
plt.hist(b,bins=[1,3,6,10,15,20],edgecolor="black")
plt.axis([0,25,0,10])
plt.show()
```
