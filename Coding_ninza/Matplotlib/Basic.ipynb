```python
x = [1,2,3]
y = [2,4,6]

x2 = [3,4,5]
y2 = [2,4,6]
plt.scatter(x,y)
plt.plot(x2,y2,"bo")
plt.plot(x,y)
plt.show()
```

```python
import numpy as np
# x = np.array([1,2,3,4])
x = np.arange(0,5,0.1)
y = x**3
y2 = x**2
plt.plot(x,y,color="black",marker="o",linewidth=5,label="x^3")
plt.plot(x,y2,color="red",marker="o",linewidth=5,label="x^2")
plt.ylabel("x^3")
plt.xlabel("x")
plt.grid()
plt.text(2,30,"text",fontsize=15)
plt.legend()         
plt.title("matplotlib")
plt.show()
```
