```python
population_India = [100,300,430,480,540,580]
population_China = [100,120,300,330,450,490]
population_Usa   = [100,106,115,125,135,150]

year = [15,16,17,18,19,20]
```

```python
plt.title('Total_Population')
plt.xlabel('year')
plt.ylabel('population')
plt.plot(year,population_India,marker="*",color='r',label='India',ls="dashed",lw=3)
plt.plot(year,population_China,marker="*",color='y',label='China',ls="dotted",lw=3)
plt.plot(year,population_Usa,marker="*",color='b',label='Usa',lw=3)
plt.legend()
plt.show()
```
